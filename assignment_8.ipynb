# Data Wrangling

Perform data wrangling.  You are free to use your best judgment here.  If you are stuck, look at previous assignment.

I am going to select a few columns of the public database to analize the data on it. Also I will create plots using Seaborn and Matplotlib to visualize the analisys.

```python
# Select columns to subset the data, I chose 9 columns to work with.
nyc_newdata =  nyc_data[["DBA","BORO","STREET","ZIPCODE","Latitude","Longitude","Community Board","Council District","Census Tract"]]
nyc_newdata.head()
```

```python
#Rename some of the columns for a better understanding of tittles.
nyc_newdata = nyc_newdata.rename(columns=  {"DBA": "Business_Name", "ZIPCODE": "ZCode", "Census Tract": "Cencus"})
print(nyc_newdata)
```

```python
# Display data types of each column
print(( nyc_newdata.dtypes))
```

```python
#Fill missing values using the NA command.
nyc_newdata.fillna(0) 
print(nyc_newdata)
```

```python
#Create a new column using the sum of columns "community board and council district"and dispaly results in a new one.
nyc_newdata['Total_Score'] = nyc_newdata['Community Board'] + nyc_newdata['Council District']
print(nyc_newdata)
```

```python
#Capitalize "BORO" column using str.upper function.
nyc_newdata['BORO'] = nyc_newdata['BORO'].str.upper() 
nyc_newdata
```

```python
#Get the mean, min, and max by grouping Counsil District, and Total Score.
nyc_newdata1 = nyc_newdata.groupby('Council District').agg({'Total_Score': ['mean', 'min', 'max']}) 
nyc_newdata1
```

```python
# Create a new Function to work with plots
nyc_latitude_mean = nyc_newdata \
    .groupby('Latitude') \
    .mean() \
    .reset_index()

nyc_latitude_mean.head(6)
```

```python
# Extract the Community Board and Total Score columns
board = nyc_latitude_mean['Community Board']
total = nyc_latitude_mean['Total_Score']
# Create a line plot
plt.plot(board, color = 'green', linewidth = 1, linestyle='dotted', marker='x')
plt.plot(total,color = 'blue', linewidth = 1,linestyle= 'dashed', marker='o')
# Label X and Y axis
plt.xlabel('Board')
plt.ylabel('Total Score')
# Adding title to plot
plt.title('NYC Restaurants Community board')
#Add anottations to plot
plt.text(3.01,3, 'Results of inspections')
# Adding a legend display in a upper left side, new fontsize, outside the plot
plt.legend(["Community", "Total Score"], loc="upper left", fontsize="15", bbox_to_anchor=(1.01, 1))
# Adding grid to plot
plt.grid(True)
# display the plot
plt.show()
```

```python

# Create a Scatter plot
plt.scatter(nyc_newdata['Community Board'], nyc_newdata['Cencus'], cmap = plt.cm.plasma)
# Label X and Y axis
plt.xlabel('Community Board')
plt.ylabel('Cencus')
# Adding title to plot
plt.title('Community B vs Cencus')
# Adding a legend display in a upper left side, new fontsize, outside the plot
plt.legend(["Community", "Cencus"], loc="lower right", fontsize="10", bbox_to_anchor=(1.01, 1))
# display the plot
plt.show()
```
