```python
#Subset the data set, I chose 7 columns to work with.
epl_subset =  epl_data[["HomeTeam","AwayTeam","FTR","HTR","HST","AST","AS"]]
epl_subset.head()
```

```python
#Rename columns for a better understanding of each one.
epl_subset1 = epl_subset.rename(columns=  {"HomeTeam": "HTeam", "AwayTeam": "ATeam", "FTR": "FullTime_Res","HTR":"HalfTime_Res","HST":"HTeam_shotson_target", "AST": "ATeam_shotson_target", "AS":"ATeam_shots"})
print(epl_subset1)
```

```python
#Print data types of columns.
print(( epl_subset1.dtypes))
```

```python
#Fill missing values using the NA command.
epl_subset1.fillna(0) 
print(epl_subset)
```

```python
#Create a new column using the sum of 2 columns and putting the result and a new one.
epl_subset1['Total_Shots'] = epl_subset1['HTeam_shotson_target'] + epl_subset1['ATeam_shotson_target']
print(epl_subset1)
```

```python
#Drop "HalfTime_Res" column from dataframe.
epl_subset1.pop("HalfTime_Res")
epl_subset1
```

```python
#Remove row #3 from dataframe.
epl_subset2= epl_subset1.drop(3)
epl_subset2
```

```python
#Sort values from the "ATeam_shots" in descending order.
epl_subset3 = epl_subset2.sort_values(by=["ATeam_shots"], ascending=False)
epl_subset3
```

```python
#Filter Data from two columns with a condition.
epl_subset4 = epl_subset3[(epl_subset3["HTeam_shotson_target"]>=5) & (epl_subset3['HTeam'].str.contains('Man City','Liverpool'))]
epl_subset4                      
```

```python
#Capitalize "HTeam" column using str.upper function.
epl_subset4['HTeam'] = epl_subset4['HTeam'].str.upper() 
epl_subset4
```

```python
#Check if data in "ATeam_shots" column is numeric, using the dtype.kind function.
epl_subset4['ATeam_shots'].dtype.kind in 'iufc'
```

```python
#Get the mean, min, and max by grouping HTeam, and Total shots column.
epl_subset5 = epl_subset4.groupby('HTeam').agg({'Total_Shots': ['mean', 'min', 'max']}) 
epl_subset5
```

```python
#Grouped two columns and print sorted results in a new column("Combined_ATeam_shots")
epl_subset4['Combined_ATeam_shots'] = epl_subset4['ATeam_shotson_target']+ epl_subset4['ATeam_shots']
print(epl_subset4.sort_values(by=['Combined_ATeam_shots'],ascending=False))
```
